###  `[25 pts]` Punto 1: Hacer análisis de sentimiento de los datos del anterior taller
- Columna con el puntaje numérico del sentimiento
- Pueden usar cualquiera de las librerias vistas en clase

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

import tweepy 
 
vader = SentimentIntensityAnalyzer()
```

```python
## Leer llaves 
keys = pd.read_csv("keys2.csv", sep = ";", header=None)
keys = dict(zip(keys[0],keys[1]))

# Autorizar a traves API 
auth = tweepy.OAuthHandler(keys['api_key'], keys['api_key_secret'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)

tweets = tweepy.Cursor(api.search, q=['covid -RT'], lang = 'es',
                      geocode = f'9.304744,-75.397778,45mi',
                      tweet_mode='extended',
                      include_rts=False).items(100)

```

```python
t = []
 
for tweet in tweets:
     temp = []
     temp.append(tweet.full_text)
     temp.append(tweet.created_at)
     temp.append("Bogota")
     temp.append(tweet.user.location)
     temp.append(tweet.id)
     temp.append(f"https://twitter.com/i/web/status/{tweet.id}")
     t.append(temp)
```

```python
data = pd.DataFrame(t)

data.columns = ['tweet', 'date', 'city', 'user_location', 'id', 'link']

data.head()
```

```python
from googletrans import Translator
translator = Translator()

data['ing'] = data['tweet'].apply(lambda valor: translator.translate(valor, src='es', dest='en'))

data.head()
```

```python
data['ingles'] = data['ing'].apply(lambda valor: valor.text)

data.head()
```

```python
data['sentimiento'] = data['ingles'].apply(lambda valor: vader.polarity_scores(valor)['compound'])

data.head()
```

```python
data[['tweet', 'sentimiento']].sort_values('sentimiento')
```

###  `[10 pts]` Punto 2: ¿Cuál es el tuit más positivo y más negativo?

```python
#Mas negativo
data[['tweet', 'sentimiento']].sort_values('sentimiento').head(1)
print(data['tweet'].loc[65])
```

El tweet mas negativo fue el numero 65 y como se puede apreciar en el texto usa palabras bastante negativas. 

```python
#Mas positivo
data[['tweet', 'sentimiento']].sort_values('sentimiento').tail(1)

print(data['tweet'].loc[87])
```

El tweet mas positivo fue el numero 87 donde alguien dice que se gano la loteria. 

###  `[15 pts]` Punto 3: ¿Cuántos tuits positivos/negativos hay en sus datos?

```python
import numpy as np
data['Status'] = np.where(data['sentimiento']>=0, 1, 0)
data.head()
```

```python
data[['Status']].sum()
```
