```python
# Abrir archivo en Python
with open("UNA_SOLITARIA_VOZ_HUMANA.txt", encoding="utf-8") as archivo:
    data = archivo.read()

print(data)
```

```python
#Minusculas 

data = data.lower() #Minuscula
print(data)
```

```python
import re
data = re.sub(r"[\W\d_]+"," ", data) #Dejar solo letras

print(data)
```

```python
data = data.split() #Tokenización
print(data)
```

# Punto 2:

- `[8pts]` Quitar palabras vacias
- `[9pts]` ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import nltk #Una vez
nltk.download('stopwords') #Una vez 
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
data = [palabra for palabra in data if palabra not in stopwords_sp]

data
```

```python
from collections import Counter

top_10 = Counter(data).most_common()

top_10[0:10]
```

# Punto 3:

- `[8pts]` Stemming del documento
- `[9pts]` ¿Cuáles son las 10 raíces más usadas?

```python
from nltk.stem.snowball import SnowballStemmer
englishStemmer=SnowballStemmer("english")
spanishStemmer=SnowballStemmer("spanish")
```

```python
texto_stem = []
 
for palabra in data:

     raiz = spanishStemmer.stem(palabra)

     texto_stem.append(raiz)


texto_stem
```

```python
from collections import Counter

top_10_raices = Counter(texto_stem).most_common()
top_10_raices[0:10]
```
